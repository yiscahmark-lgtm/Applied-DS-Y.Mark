```python
import PyPDF2
```

Extracting text from a text file

```python
with open('Week_3_No_Tables.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    text = reader.pages[0].extract_text()
    print("--- Simple PDF Results ---")
    print(text)
```

Extracting text from a table

```python
import pdfplumber


with pdfplumber.open('Week_3_With_Tables.pdf') as pdf:
    first_page = pdf.pages[0]
    table = first_page.extract_table()
    print("--- Table PDF Results ---")
    for row in table:
        print(row)
```
